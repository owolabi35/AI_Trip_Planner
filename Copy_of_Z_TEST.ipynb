A factory produces ball of a diameter of 5cm. But due to manufacturing conditions, every ball may not have the exactly same diameter.The standard dev. of the diameters of balls is 0.4. Now, the quality officer would like to test whether the ball diamter is significantly different from 5cm in a sample of 50 balls randomly taken from the manufacturing line

null hyp. mean = 5
alternate mean not equal 5

```python
res = stat()
res.ztest(df=df, x='sizes', mu=5, alpha=0.05, x_std=0.4, test_type=1)
```

```python
print(res.summary)
```

df: pandas dataframe containg x group
x: column name for group
mu: hypothsized or known popn mean(specified in null hypothesis)
x_std: popn std dev for x group
alpha: significance level for C.I (default=0.05)
test_type: Type of Z-test. Use 1 for one sample Z-test
